```python
## Making buckets with respect to stays in week nights

night_bins = [0,6,11,16,21]
night_labels = ['0-5','6-10','11-15','16-20']

df_hotel['stay_week_nights_bucket'] = pd.cut(df_hotel['stays_in_week_nights'],bins= night_bins,labels=night_labels,
                                             right=False,include_lowest=True)
df_hotel['stay_week_nights_bucket'] = df_hotel['stay_week_nights_bucket'].astype('str')
df_hotel['stay_week_nights_bucket'] = df_hotel['stay_week_nights_bucket'].replace({'nan':'> 20'})
df_hotel['stay_week_nights_bucket'].value_counts()
```

```python
lead_bins = [0,6,11,21,31,41,51,76,101,126,151,176,201,251,301,351,401,451,501]
lead_labels = ['A. 0-5','B. 6-10','C. 11-20','D. 21-30','E. 31-40','F. 41-50','G. 51-75','H. 76-100','I. 101-125',
               'J. 126-150','K. 151-175','L. 176-200','M. 201-250','N. 251-300','O. 301-350','P. 351-400',
               'Q. 401-450','R. 451-500']
               


df_hotel['lead_bucket'] = pd.cut(df_hotel['lead_time'],bins= lead_bins,labels=lead_labels,
                                             right=False,include_lowest=True)
df_hotel['lead_bucket'] = df_hotel['lead_bucket'].astype('str')
df_hotel['lead_bucket'] = df_hotel['lead_bucket'].replace({'nan':'S. > 500'})
df_hotel['lead_bucket'].value_counts().sort_index()
```

```python
## Combining the year, month, day into a single date column


month_dict = {'January':1,'February':2,'March':3,'April':4,'May':5,'June':6,'July':7,'August':8,'September':9,
              'October':10,'November':11,'December':12}
df_hotel['arrival_month'] = df_hotel['arrival_date_month']
df_hotel['arrival_month'] = df_hotel['arrival_month'].replace(month_dict)


df_hotel['arrival_date'] =df_hotel['arrival_date_year'].astype('str') + "-" + df_hotel['arrival_month'].astype('str') + \
 "-" + df_hotel['arrival_date_day_of_month'].astype('str')   
df_hotel['arrival_date'] = pd.to_datetime(df_hotel['arrival_date'])

df_hotel[['arrival_date','arrival_date_year','arrival_date_month','arrival_date_day_of_month']]
```
